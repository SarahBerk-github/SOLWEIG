```python
import pandas as pd
import numpy as np

###########################################################
# Load ERA5 city-mean forcing from GEE
###########################################################

era5_file = r"C:\Users\Sarah\Documents\Python_Scripts\DDL\SOLWEIG\Data\ERA5_Durham_city_mean_hourly.csv"

df = pd.read_csv(era5_file)

###########################################################
# Convert time
###########################################################

# get the date from system:index 

df['datetime'] = pd.to_datetime(df['system:index'])

df['%iy'] = df['datetime'].dt.year
df['it']  = df['datetime'].dt.hour
df['imin'] = df['datetime'].dt.minute

###########################################################
# Convert units
###########################################################

# Temperature from K to C 
df['Td'] = df['temperature_2m'] - 273.15

# Relative humidity
df['RH'] = df['relative_humidity']

# Radiation: J/m2/hr → W/m2
df['Kdn'] = df['surface_solar_radiation_downwards_hourly'] / 3600
df['ldown'] = df['surface_thermal_radiation_downwards_hourly'] / 3600

###########################################################
# Fill required SOLWEIG columns
###########################################################

df['id'] = 214

missing_cols = [
    'Q*','QH','QE','Qs','Qf',
    'Wind','press','rain',
    'snow','fcld','wuh',
    'xsmd','lai_hr',
    'Kdiff','Kdir','Wd'
]

for col in missing_cols:
    df[col] = -999.00

###########################################################
# Keep only required column order
###########################################################

columns_order = [
    '%iy','id','it','imin',
    'Q*','QH','QE','Qs','Qf',
    'Wind','RH','Td','press',
    'rain','Kdn','snow',
    'ldown','fcld','wuh',
    'xsmd','lai_hr',
    'Kdiff','Kdir','Wd'
]

df_out = df[columns_order]

###########################################################
# Export SOLWEIG met file
###########################################################

output_file = r"C:\Users\Sarah\Documents\Python_Scripts\DDL\SOLWEIG\Data\Durham_ERA5_SOLWEIG_met.txt"

df_out.to_csv(
    output_file,
    sep=' ',
    index=False,
    float_format="%.2f"
)

print("SOLWEIG met file created:")
print(output_file)
```
